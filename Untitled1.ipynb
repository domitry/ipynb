```python
%matplotlib inline
```

```python
from ipysurfer.mri import CategorizedMRI
from ipysurfer.widget import Widget
```

```python
mri = CategorizedMRI.from_mgz("/home/naoki/data/brain/aparc.a2009s+aseg2.mgz", "/home/naoki/data/brain/ColorLUT.txt")
```

```python
wid = Widget("/home/naoki/data/brain/testDT3.mgz", mri)
wid.with_movie("https://dl.dropbox.com/s/jmwmt26c159dfhg/output-file.webm")
```

```python
wid.plot()
```
